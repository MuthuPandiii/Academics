```python
import pandas as pd
import os
import numpy as np
```

```python
file_path = os.path.join(os.getcwd(),'Data')

files = os.listdir(os.path.join(os.getcwd(),'Data'))

data = {'Name':[],'Roll Number':[],'Semester':[],'Course':[],
        '19CSBS401-DISCRETE MATHEMATICS':[],
        '19CSCP407-DATABASE MANAGEMENT SYSTEMS LAB':[],
        '19CSCP408-OPERATING SYSTEMS LAB':[],
        '19CSCP409-PYTHON PROGRAMMING LAB':[],
        '19CSES402-DESIGN AND ANALYSIS OF ALGORITHMS':[],
        '19CSPC403-DATABASE MANAGEMENT SYSTEMS':[],
        '19CSPC404-OPERATING SYSTEMS':[],
        '19CSPC405-PYTHON PROGRAMMING':[],
        '19CSPC406-COMPUTER ORGANIZATION AND ARCHITECTURE':[],
        'Month/Year':[]}

subjects = ['19CSBS401-DISCRETE MATHEMATICS',
        '19CSCP407-DATABASE MANAGEMENT SYSTEMS LAB',
        '19CSCP408-OPERATING SYSTEMS LAB',
        '19CSCP409-PYTHON PROGRAMMING LAB',
        '19CSES402-DESIGN AND ANALYSIS OF ALGORITHMS',
        '19CSPC403-DATABASE MANAGEMENT SYSTEMS',
        '19CSPC404-OPERATING SYSTEMS',
        '19CSPC405-PYTHON PROGRAMMING',
        '19CSPC406-COMPUTER ORGANIZATION AND ARCHITECTURE',]
```

```python
for i in range(len(files)):
    raw_data = []
    for line in open(os.path.join(file_path,files[i]),'r'):
        raw_data.append(line.split())
    details = raw_data[:3]
    score = raw_data[4:]

    data['Roll Number'].append(details[0][-1])
    data['Name'].append(' '.join(details[1][2:]))
    data['Course'].append(' '.join(details[2][2:]))
    data['Semester'].append(score[0][0])
    data['Month/Year'].append(score[0][-1])

    for j in range(len(subjects)):
        data[subjects[j]].append(score[j][-4])
    print(f'{files[i]} completed')
```

```python
df = pd.DataFrame(data)
df.index = np.arange(1,len(df)+1)
```
